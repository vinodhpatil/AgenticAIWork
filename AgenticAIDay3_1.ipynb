```python
import os
from google.colab import userdata
from datetime import datetime
import pytz

import google.generativeai as genai

# Configure API key
genai.configure(api_key=userdata.get("GOOGLE_API_KEY"))

# Discover available models
available_models = []
for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        available_models.append(m.name)

print("Available models:", available_models)

# Select first available model
if not available_models:
    raise Exception("No compatible Gemini models available for this API key")

model_name = available_models[0]
print("Using model:", model_name)

#STEP 2 : setup gemini model

model = genai.GenerativeModel(model_name)

#Agent 1: symptom Agent
def symptom_agent(input_text):
  print("\n[Symptom Agent]")
  response = model.generate_content(f"Classify the severity and list possible conditions:{input_text}")
  print(response.text)
  return response.text


#Agent 2: Scheduler Agent
def scheduler_agent(symptom_severity_summary):
  print("\n[Scheduler Agent]")
  response = model.generate_content(f"Based on symptom summary, suggest next steps for patient:{symptom_severity_summary}")
  print(response.text)
  return response.text

chat = model.start_chat()

#Agent 3: Followup agent
def followup_agent():
  print("\n[Followup Agent]")
  response = model.generate_content(f"What follow up question should doctor ask to understand patient condition")
  print(response.text)
  return response.text

#Agent 4: Logging Agent
def logging_agent(name,pid, input_text, output_text):
  print("\n[Logging Agent]")
  log_entry = f"""\nTime:{datetime.now(pytz.timezone('Asia/Kolkata')).strftime('%d-%m-%Y %H:%M:%S')}\n
  Patient: {name}\n
  PID: {pid}\n
  Input: {input_text}\n
  AI Summary: {output_text}\n
  {'='*50}\n"""
  with open("prescription_conversation.txt", "a") as f:
    f.write(log_entry)

#Define user inputs
name = input("Please enter patient name: ")
pid = input("Please enter PID: ")
user_input = input("Please describe your health symptoms :")

# Agent 1 calling
symptom_severity_summary = symptom_agent(user_input)

# Agent 2 calling
next_steps = scheduler_agent(symptom_severity_summary)

# Agent 3 calling
followup_question = followup_agent()

# Agent 4 calling
logging_agent(name,pid, user_input, symptom_severity_summary)
```
